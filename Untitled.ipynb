```python
!git clone https://github.com/engmahsa/Senti-Persian-Dataset.git

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from collections import Counter
from wordcloud import WordCloud
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import arabic_reshaper
from bidi.algorithm import get_display
from google.colab import files
```

```python
df = pd.read_excel('Senti-Persian-Dataset/data/Sentiment Analysis data.xlsx', header=None)
df.columns = df.iloc[1]
df = df.drop([0,1]).reset_index(drop=True)
df['label'] = df['label'].astype(int)
df['comment'] = df['comment'].astype(str)
print(df.head())
```

```python
print("ŸÑÿ∑ŸÅÿßŸã ŸÅÿß€åŸÑ ŸÅŸàŸÜÿ™ (TTF €åÿß OTF) ÿ±Ÿà ÿ¢ŸæŸÑŸàÿØ ⁄©ŸÜ")
uploaded = files.upload()
font_file = None
for filename in uploaded.keys():
    if filename.lower().endswith(('.ttf', '.otf')):
        font_file = filename
        break
if font_file is None:
    raise FileNotFoundError('Ÿá€å⁄Ü ŸÅÿß€åŸÑ ŸÅŸàŸÜÿ™€å ÿ¢ŸæŸÑŸàÿØ ŸÜÿ¥ÿØŸá!')
font_path = font_file
font_prop = fm.FontProperties(fname=font_path)
plt.rcParams['font.family'] = font_prop.get_name()

# ŸÜŸÖÿß€åÿ¥ ÿ™ÿ≥ÿ™ ŸÅŸàŸÜÿ™
text = 'ÿ≥ŸÑÿßŸÖ ÿØŸÜ€åÿßÿå ÿß€åŸÜ €å⁄© ÿ™ÿ≥ÿ™ ŸÅŸàŸÜÿ™ ŸÅÿßÿ±ÿ≥€å ÿßÿ≥ÿ™'
reshaped_text = arabic_reshaper.reshape(text)
bidi_text = get_display(reshaped_text)
plt.figure(figsize=(8,4))
plt.text(0.5, 0.5, bidi_text, fontproperties=font_prop, fontsize=28, ha='center', va='center')
plt.axis('off')
plt.show()
```

```python
stopwords_fa = ['Ÿà', 'ÿØÿ±', 'ÿ®Ÿá', 'ÿßÿ≤', '⁄©Ÿá', 'ÿ®ÿß', 'ÿ®ÿ±ÿß€å', 'ÿßÿ≥ÿ™', 'ÿß€åŸÜ', 'ÿ¢ŸÜ', 'ÿ±ÿß', 'ŸÖ€å']

def normalize_fa(text):
    text = text.replace('Ÿä', '€å').replace('ŸÉ', '⁄©')
    text = re.sub(r'[€∞-€π0-9]', '', text)  # ÿ≠ÿ∞ŸÅ ÿßÿπÿØÿßÿØ
    text = re.sub(r'[^\w\s]', '', text)    # ÿ≠ÿ∞ŸÅ ÿπŸÑÿßÿ¶ŸÖ ŸÜ⁄Øÿßÿ±ÿ¥€å
    text = re.sub(r'\s+', ' ', text)       # ÿ≠ÿ∞ŸÅ ŸÅÿßÿµŸÑŸá‚ÄåŸáÿß€å ÿßÿ∂ÿßŸÅ€å
    return text.strip()

def tokenize_fa(text):
    words = text.split()
    words = [w for w in words if w not in stopwords_fa]
    return words

df['processed_comment'] = df['comment'].apply(lambda x: tokenize_fa(normalize_fa(x)))
```

```python
df['processed_text'] = df['processed_comment'].apply(lambda x: ' '.join(x))
X = df['processed_text']
y = df['label']

vectorizer = TfidfVectorizer(max_features=5000)
X_tfidf = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import string

# Drop the previous lowercase column if it exists
if "text_lower" in df.columns:
    df.drop(["text_lower"], axis=1, inplace=True)

# Define a function to remove punctuation from text
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """Remove punctuation from a string"""
    if isinstance(text, str):
        return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))
    return text

# Apply the function and create a new column without punctuation
df["text_no_punct"] = df["processed_text"].apply(remove_punctuation)

# Display the first few rows to check the result
df.head()
```

```python
# Custom simple stemmer for Persian without Hazm
def simple_persian_stemmer(word):
    # Common Persian suffixes
    suffixes = ["Ÿáÿß", "Ÿáÿß€å", "ÿ™ÿ±", "ÿ™ÿ±€åŸÜ", "ÿßŸÖ", "ÿßÿ™", "ÿßÿ¥", "ŸÖÿßŸÜ", "ÿ™ÿßŸÜ", "ÿ¥ÿßŸÜ"]
    for suf in suffixes:
        if word.endswith(suf):
            return word[:-len(suf)]
    return word

# Apply to your processed_text column
df["text_stemmed"] = df["processed_text"].apply(
    lambda text: " ".join([simple_persian_stemmer(w) for w in text.split()])
)

# Show some results
print(df[["processed_text", "text_stemmed"]].head())
```
