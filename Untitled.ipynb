```python
!git clone https://github.com/engmahsa/Senti-Persian-Dataset.git

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from collections import Counter
from wordcloud import WordCloud
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import arabic_reshaper
from bidi.algorithm import get_display
from google.colab import files
```

```python
df = pd.read_excel('Senti-Persian-Dataset/data/Sentiment Analysis data.xlsx', header=None)
df.columns = df.iloc[1]
df = df.drop([0,1]).reset_index(drop=True)
df['label'] = df['label'].astype(int)
df['comment'] = df['comment'].astype(str)
print(df.head())
```

```python
print("لطفاً فایل فونت (TTF یا OTF) رو آپلود کن")
uploaded = files.upload()
font_file = None
for filename in uploaded.keys():
    if filename.lower().endswith(('.ttf', '.otf')):
        font_file = filename
        break
if font_file is None:
    raise FileNotFoundError('هیچ فایل فونتی آپلود نشده!')
font_path = font_file
font_prop = fm.FontProperties(fname=font_path)
plt.rcParams['font.family'] = font_prop.get_name()

# نمایش تست فونت
text = 'سلام دنیا، این یک تست فونت فارسی است'
reshaped_text = arabic_reshaper.reshape(text)
bidi_text = get_display(reshaped_text)
plt.figure(figsize=(8,4))
plt.text(0.5, 0.5, bidi_text, fontproperties=font_prop, fontsize=28, ha='center', va='center')
plt.axis('off')
plt.show()
```

```python
stopwords_fa = ['و', 'در', 'به', 'از', 'که', 'با', 'برای', 'است', 'این', 'آن', 'را', 'می']

def normalize_fa(text):
    text = text.replace('ي', 'ی').replace('ك', 'ک')
    text = re.sub(r'[۰-۹0-9]', '', text)  # حذف اعداد
    text = re.sub(r'[^\w\s]', '', text)    # حذف علائم نگارشی
    text = re.sub(r'\s+', ' ', text)       # حذف فاصله‌های اضافی
    return text.strip()

def tokenize_fa(text):
    words = text.split()
    words = [w for w in words if w not in stopwords_fa]
    return words

df['processed_comment'] = df['comment'].apply(lambda x: tokenize_fa(normalize_fa(x)))
```

```python
df['processed_text'] = df['processed_comment'].apply(lambda x: ' '.join(x))
X = df['processed_text']
y = df['label']

vectorizer = TfidfVectorizer(max_features=5000)
X_tfidf = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
df_test = pd.DataFrame({
    'comment': df.loc[y_test.index, 'processed_text'],  # استفاده از متن پردازش شده
    'true_label': y_test.values,
    'pred_label': y_pred
})

errors = df_test[df_test['true_label'] != df_test['pred_label']]

# ترکیب تمام کلمات نمونه‌های اشتباه
all_error_words = ' '.join(errors['comment'])

wordcloud_errors = WordCloud(
    font_path=font_path,
    width=800,
    height=400,
    background_color='white'
).generate(all_error_words)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud_errors, interpolation='bilinear')
plt.axis('off')
plt.title("کلمات پرتکرار در نمونه‌های اشتباه")
plt.show()
```

```python
import string

# Drop the previous lowercase column if it exists
if "text_lower" in df.columns:
    df.drop(["text_lower"], axis=1, inplace=True)

# Define a function to remove punctuation from text
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """Remove punctuation from a string"""
    if isinstance(text, str):
        return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))
    return text

# Apply the function and create a new column without punctuation
df["text_no_punct"] = df["processed_text"].apply(remove_punctuation)

# Display the first few rows to check the result
df.head()
```

```python
# Custom simple stemmer for Persian without Hazm
def simple_persian_stemmer(word):
    # Common Persian suffixes
    suffixes = ["ها", "های", "تر", "ترین", "ام", "ات", "اش", "مان", "تان", "شان"]
    for suf in suffixes:
        if word.endswith(suf):
            return word[:-len(suf)]
    return word

# Apply to your processed_text column
df["text_stemmed"] = df["processed_text"].apply(
    lambda text: " ".join([simple_persian_stemmer(w) for w in text.split()])
)

# Show some results
print(df[["processed_text", "text_stemmed"]].head())
```
